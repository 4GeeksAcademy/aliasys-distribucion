# Problemas de distribución de probabilidad

```python
# Importar las librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
from scipy.stats import norm
```

## Distribución binomial

En teoría de la probabilidad y estadística, la distribución binomial es la distribución de probabilidad discreta que da sólo dos resultados posibles en un experimento; por ejemplo, al lanzar una moneda, puede salir cara o cruz.

```python
# Definir la función de visualización de la distribución binomial
# Definir todas las funciones de probabilidad relacionadas con esta distribución.

def dbinom(x, size, prob = 0.5):
    """
    Calcula la estimación puntual de la distribución binomial.
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calcula el acumulado de la distribución binomial.
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calcula la función cuantil a partir de la distribución binomial.
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Genera variables aleatorias a partir de la distribución binomial.
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Distribución binomial(10, 0.2), dónde n = 10 es el número de ensayos y p = 0.2 es la probabilidad de éxito en un solo ensayo

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
# Visualización de la distribución binomial(10, 0.2)
np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

# Distribucion normal

```python
# Empezamos con las distribuciones normales
def dnorm(x, mean = 0, sd = 1):
    """
    Calcula la estimación puntual de la distribución normal.
    """
    result = norm.pdf(x, loc = mean, scale = sd)
    return result
def pnorm(q, mean = 0, sd = 1):
    """
    Calcula el acumulado de la distribución normal.
    """
    result = norm.cdf(q, loc = mean, scale = sd)
    return result
def qnorm(p, mean = 0, sd = 1):
    """
    Calcula la función cuantil a partir de la distribución normal.
    """
    result = norm.ppf(p, loc = mean, scale = sd)
    return result
def rnorm(n, mean = 0, sd = 1):
    """
    Genera variables aleatorias a partir de la distribución normal.
    """
    result = norm.rvs(loc = mean, scale = sd, size = n)
    return result
```

```python
# Distribución normal(0, 1), dónde 0 es la media y 1 es la desviación estándar
np.random.seed(42)
print(f"Probability that a Normal(0, 1) takes the value 2: {dnorm(2, mean = 0, sd = 1)}")
print(f"Probability that a Normal(0, 1) will take a value less than 2: {pnorm(2, mean = 0, sd = 1)}")
print(f"Which value of a Normal(0, 1) has a cumulative probability of 0.9? {qnorm(0.9, mean = 0, sd = 1)}")
print(f"Generate 2000 random values from a Normal(0, 1) distribution: {rnorm(2000, mean = 0, sd = 1)}")
```

```python
#Visualización de la distribución normal(0, 1)
np.random.seed(42)
mean = 0
sd = 1
data = rnorm(2000, mean = mean, sd = sd)
pmf = dnorm(data, mean = mean, sd = sd)
cdf = pnorm(data, mean = mean, sd = sd)
ppf = qnorm(data, mean = mean, sd = sd)
fig, axis = plt.subplots(3, 1, figsize = (5, 7))
axis[0].plot(data, pmf, "r-")
axis[1].plot(data, cdf, "b-")
axis[2].plot(data, ppf, "g-")
plt.tight_layout()
plt.show()
```

# Distribucion uniforme

```python
# Calculamos la distribucion uniforme
def dunif(x, min = 0, max = 1):
    """
    Calcula la estimación puntual de la distribución uniforme.
    """
    if min <= x <= max:
        return 1 / (max - min)
    else:
        return 0
def punif(q, min = 0, max = 1):
    """
    Calcula el acumulado de la distribución uniforme.
    """
    if q < min:
        return 0
    elif q > max:
        return 1
    else:
        return (q - min) / (max - min)
def qunif(p, min = 0, max = 1):
    """
    Calcula la función cuantil a partir de la distribución uniforme.
    """
    if p < 0 or p > 1:
        raise ValueError("p must be between 0 and 1")
    return min + p * (max - min)
def runif(n, min = 0, max = 1):
    """
    Genera variables aleatorias a partir de la distribución uniforme.
    """
    return np.random.uniform(low = min, high = max, size = n)
```

```python
# Distribucuión uniforme(0, 1), dónde 0 es el mínimo y 1 es el máximo
np.random.seed(42)
print(f"Probability that a Uniform(0, 1) takes the value 0.5: {dunif(0.5, min = 0, max = 1)}")
print(f"Probability that a Uniform(0, 1) will take a value less than 0.5: {punif(0.5, min = 0, max = 1)}")
print(f"Which value of a Uniform(0, 1) has a cumulative probability of 0.9? {qunif(0.9, min = 0, max = 1)}")
print(f"Generate 2000 random values from a Uniform(0, 1) distribution: {runif(2000, min = 0, max = 1)}")
```

```python
# Visualización de la distribución uniforme(0, 1)
np.random.seed(42)
min_val = 0
max_val = 1
data = runif(2000, min = min_val, max = max_val)
pmf = [dunif(x, min = min_val, max = max_val) for x in data]
cdf = [punif(q, min = min_val, max = max_val) for q in data]
ppf = [qunif(p, min = min_val, max = max_val) for p in np.linspace(0, 1, 2000)]
fig, axis = plt.subplots(3, 1, figsize = (5, 7))
axis[0].plot(data, pmf, "r-")
axis[1].plot(data, cdf, "b-")
axis[2].plot(data, ppf, "g-")
plt.tight_layout()
plt.show()
```
